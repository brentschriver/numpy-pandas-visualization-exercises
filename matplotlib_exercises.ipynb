```python
import matplotlib.pyplot as plt
```

### 1. Use matplotlib to plot the following equation:
#### y = x^2 - x + 2
#### You'll need to write the code that generates the x and y points.
#### Add an anotation for the point 0, 0, the origin.

```python
x = range(10)
y = [num**2 - num + 2 for num in x]

plt.plot(x,y)
plt.annotate('Origin (0,0)', xy=(0, 0), xytext=(0, 10))
plt.show()
```

### 2. Create and label 4 separate charts for the following equations (choose a range for x that makes sense):
 * y = x^(1/2)
 * y = x^3
 * y = 2^x
 * y = 1/(x+1)
 
#### You can use functions from the math module to help implement some of the equations above.

```python
# Create a range of values from 0 to 10.
x = range(10)
# For every number in x, take the square root.
y1 = [num ** (1/2) for num in x]
y2 = [num ** 3 for num in x]
y3 = [2 ** num for num in x]
y4 = [1/(num + 1) for num in x]

plt.subplot(221)
plt.plot(x,y1)
plt.title('$x^1/2$')

plt.subplot(222)
plt.plot(x,y2)
plt.title('$x^3$')

plt.subplot(223)
plt.plot(x,y3)
plt.title('$2^x$')

plt.subplot(224)
plt.plot(x,y4)
plt.title('$\\frac{1}{x+1}$')
```

### 3. Combine the figures you created in the last step into one large figure with 4 subplots.

```python
x = range(10)
y1 = [num ** (1/2) for num in x]
y2 = [num ** 3 for num in x]
y3 = [2 ** num for num in x]
y4 = [1/(num + 1) for num in x]

plt.figure(figsize = (14,14))

plt.subplot(2,2,1)
plt.plot(x,y1)
plt.title('x^(1/2)')

plt.subplot(2,2,2)
plt.plot(x,y2)
plt.title('$x^3$')

plt.subplot(2,2,3)
plt.plot(x,y3)
plt.title('$2^x$')

plt.subplot(2,2,4)
plt.plot(x,y4)
plt.title('$\\frac{1}{x+1}$')
```

### 4. Combine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. Be sure to include a legend and an appropriate title for the figure.

```python
x = range(10)
y1 = [num ** (1/2) for num in x]
y2 = [num ** 3 for num in x]
y3 = [2 ** num for num in x]
y4 = [1/(num + 1) for num in x]

plt.plot(x,y1)
plt.scatter(x,y1, color = 'r', label = '$\sqrt{x}$')
plt.plot(x,y2)
plt.scatter(x,y2, color = 'b', label = '$x^3$')
plt.plot(x,y3)
plt.scatter(x,y3, color = 'g', label = '$2^x$')
plt.plot(x,y4)
plt.scatter(x,y4, color = 'm', label = '$\\frac{1}{x+1}$')
plt.title('Four Different Functions')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
```
