```python
import matplotlib.pyplot as plt

x = range(10)
y = [num ** 2 - num + 2 for num in x]

plt.plot(x,y)
plt.annotate('(0,0)', xy = (0,0), xytext = (0,5))
```

```python
x = range(10)
y1 = [num ** (1/2) for num in x]

plt.plot(x,y1)
plt.title('x^(1/2)')

```

```python
x = range(10)
y2 = [num ** 3 for num in x]

plt.plot(x,y2)
plt.title('$x^3$')

```

```python
x = range(10)
y3 = [2 ** num for num in x]

plt.plot(x,y3)
plt.title('$2^x$')

```

```python
x = range(10)
y4 = [1/(num + 1) for num in x]

plt.plot(x,y4)
plt.title('1/(x+1)')
```

```python
x = range(10)
y1 = [num ** (1/2) for num in x]
y2 = [num ** 3 for num in x]
y3 = [2 ** num for num in x]
y4 = [1/(num + 1) for num in x]

plt.figure(figsize = (14,14))

plt.subplot(2,2,1)
plt.plot(x,y1)
plt.title('x^(1/2)')

plt.subplot(2,2,2)
plt.plot(x,y2)
plt.title('$x^3$')

plt.subplot(2,2,3)
plt.plot(x,y3)
plt.title('$2^x$')

plt.subplot(2,2,4)
plt.plot(x,y4)
plt.title('1/(x+1)')
```

```python
x = range(10)
y1 = [num ** (1/2) for num in x]
y2 = [num ** 3 for num in x]
y3 = [2 ** num for num in x]
y4 = [1/(num + 1) for num in x]

plt.plot(x,y1)
plt.scatter(x,y1, color = 'r', label = 'x^(1/2)')
plt.plot(x,y2)
plt.scatter(x,y2, color = 'b', label = '$x^3$')
plt.plot(x,y3)
plt.scatter(x,y3, color = 'g', label = '$2^x$')
plt.plot(x,y4)
plt.scatter(x,y4, color = 'm', label = '1/(x+1)')
plt.title('Four Different Functions')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
```
