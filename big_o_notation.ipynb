```python
import matplotlib.pyplot as plt  
import numpy as np
import math
from random import randint
import random
import pandas as pd

plt.figure(figsize = (16,16))
plt.title('Big O Notation')
plt.xlabel('Elements')
plt.ylabel('Operations')

x = range(1,10)
y1 = [num * 0 + 1 for num in x]
y2 = [math.log(num) for num in x]
y3 = [num for num in x]
y4 = [num * math.log(num) for num in x]
y5 = [num ** 2 for num in x]
y6 = [2 ** num for num in x]
y7 = [math.factorial(num) for num in x]
y8 = [num ** num for num in x]


plt.scatter(x, y1, color = 'r', label = ('O(1)'))
plt.plot(x,y1)
plt.scatter(x, y2, color = 'g', label = ('O(log n)'))
plt.plot(x,y2)
plt.scatter(x, y3, color = 'b', label = ('O(n)'))
plt.plot(x,y3)
plt.scatter(x, y4, color = 'm', label = ('O(n log n)'))
plt.plot(x,y4)
plt.scatter(x, y5, color = 'c', label = ('O($n^2$)'))
plt.plot(x,y5)
plt.scatter(x, y6, color = 'y', label = ('O($2^n$)'))
plt.plot(x,y6)
plt.scatter(x, y7, color = 'k', label = ('O(n!)'))
plt.plot(x,y7)
plt.scatter(x, y8, color = 'aquamarine', label = ('O($n^n$)'))
plt.plot(x,y8)

plt.ylim(0,100)

plt.legend()
```
