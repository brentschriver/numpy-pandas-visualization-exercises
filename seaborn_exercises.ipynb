4. Using the chipotle dataset from the previous exercise, create a bar chart that shows the 4 most popular items and the revenue produced by each.

```python
def get_db_url(db_name):
    from env import host, user, password
    return f'mysql+pymysql://{user}:{password}@{host}/{db_name}'
```

```python
chipotle_url = get_db_url('chipotle')
chipotle_sql = '''
select * from orders
'''
chipotle_dataframe = pd.read_sql(chipotle_sql, chipotle_url)
chipotle_dataframe.item_price = chipotle_dataframe.item_price.str.replace('$','').astype('float')
```

```python
most_cheese = chipotle_dataframe.groupby('item_name').item_price.agg('sum').sort_values(ascending=False).head(4)
```

```python
most_cheese = most_cheese.reset_index()

plt.figure(figsize = (8, 5))
sns.barplot(y = 'item_name', x = 'item_price', data = most_cheese)

plt.ylabel('') 
plt.xlabel('Gross Revenue')
plt.suptitle('Revenue from Top 4 Selling Items')
```
